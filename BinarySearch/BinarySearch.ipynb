```python
nums = [-1,0,3,5,9,12]
target = 9
nums2 = [-1,0,3,5,9,12]
target2 = 2
```

```python
def binarySearch(nums, target):
  i=round(len(nums)/2)
  j=len(nums)-1
  k=0
  while(nums[i]!=target and i<j and k<i):
    if nums[i]>target:
      j=i
      i=round(i-(i-k)/2)
    elif nums[i]<target:
      k=i
      i=round(i+(j-i)/2)
  if nums[i]!=target:
    return -1
  return i
```

```python
print(binarySearch(nums,target))
print(binarySearch(nums2,target2))
```
