<h1>Reverse linked list

```python
head = [1,2,3,4,5]
head2 = [1,2]
head3 = []
```

```python
def reverse(head):
  result = []
  for i in range(len(head)):
    result.append(head[i])
  return result
```

```python
print(reverse(head))
print(reverse(head2))
print(reverse(head3))
```

<h1> Merge two sorted lists

```python
list1 = [1,2,4]
list2 = [1,3,4]
list3 = []
list4 = []
list5 = []
list6 = [0]
```

```python
def merge(list1,list2):
  result = []
  i,j = 0,0
  while(len(result)!=len(list1)+len(list2)):
    if j>=len(list2) or (i< len(list1)-1 and list1[i]<=list2[j]):
      result.append(list1[i])
      i+=1
    else:
      result.append(list2[j])
      j+=1
  return result
```

```python
print(merge(list1,list2))
print(merge(list3,list4))
print(merge(list5,list6))
```
