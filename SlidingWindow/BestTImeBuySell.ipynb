```python
prices = [7,1,5,3,6,4]
prices2 = [7,6,4,3,1]
```

```python
def diffBuySell(prices):
  maxDiff=0
  for i in range(len(prices)-1):
    for j in range(i+1,len(prices)):
      if prices[j]-prices[i] > maxDiff:
        maxDiff = prices[j]-prices[i]
  return maxDiff
```

```python
print(diffBuySell(prices))
print(diffBuySell(prices2))
```
