```python
s = "()"
s2 = "()[]{}"
s3 = "(]"
```

```python
def valParenthesis(s):
  for i in range(len(s)-1):
    if s[i] == '(' and s[i+1] != ')':
      return False
    elif s[i] == '[' and s[i+1] != ']':
      return False
    elif s[i] == '{' and s[i+1] != '}':
      return False
  return True
```

```python
print(valParenthesis(s))
print(valParenthesis(s2))
print(valParenthesis(s3))
```
