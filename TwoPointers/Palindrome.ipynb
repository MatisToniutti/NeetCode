```python
s = "A man, a plan, a canal: Panama"
s2 = "race a car"
s3 = " "
```

```python
import re
def palindrome(s):
  s = re.sub(r'\W+', '', s).lower()
  print(s)
  for i in range(round(len(s)/2)):
    if(s[i]!=s[len(s)-i-1]):
      return False
  return True
```

```python
print(palindrome(s))
print(palindrome(s2))
print(palindrome(s3))
```
