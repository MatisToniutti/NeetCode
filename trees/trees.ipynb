```python
import math
def inverseTab(tab):
  for i in range(math.ceil((len(tab))/2)):
    print(tab)
    tmp = tab[i]
    tab[i] = tab[-i-1]
    tab[-i-1] = tmp
    print(tab)
  return tab
```

```python
def inverse(root):
  i=1
  while(i<len(root)):
    root[i:(i+1)*2-1] = inverseTab(root[i:(i+1)*2-1])
    i= (i+1)*2-1
  return root
```
